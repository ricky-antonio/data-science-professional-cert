# Task 2: Overfitting

Split the data set into training and testing components again, this time reserving 50% of the data set for testing.


````python
# Write your code below and press Shift+Enter to execute 

# split the dataset into training and test sets (50% test size) using a fixed random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.5, random_state=0)
````

To identify the point of overfitting the model on the parameter "CPU_frequency", you'll need to create polynomial features using the single attribute. 
You need to evaluate the R^2 scores of the model created using different degrees of polynomial features, ranging from 1 to 5. Save this set of values of R^2 score as a list.


````python
# Write your code below and press Shift+Enter to execute 

# iterate over polynomial degrees to evaluate model performance
order = [1,2,3,4,5]
Rsqu_test = []

lre = LinearRegression()

for n in order:
    # create polynomial feature transformer for the current degree
    pr = PolynomialFeatures(degree=n)

    # fit the transformer on the training data and apply it to both train and test sets
    X_train_pr = pr.fit_transform(X_train[['CPU_frequency']])
    X_test_pr = pr.transform(X_test[['CPU_frequency']])

    # train the linear regression model on the polynomial transformed training data
    lre.fit(X_train_pr, y_train)

    # compute the R^2 score on the test set and store it
    Rsqu_test.append(lre.score(X_test_pr, y_test))
````

# Task 3 : Ridge Regression

Now consider that you have multiple features, i.e. 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU' and 'Category'. Create a polynomial feature model that uses all these parameters with degree=2. Also create the training and testing attribute sets.


````python
# Write your code below and press Shift+Enter to execute 

# instatiate polynomial features model with a degree of 2
pr = PolynomialFeatures(degree=2)

# fit the transformer on the training data and transform both training and test sets
X_train_pr = pr.fit_transform(X_train[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU', 'Category']])
X_test_pr = pr.transform(X_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU', 'Category']])
````

<details><summary>Click here for the solution</summary>

```python
pr=PolynomialFeatures(degree=2)

x_test_pr=pr.fit_transform(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
```

</details>


Create a Ridge Regression model and evaluate it using values of the hyperparameter alpha ranging from 0.001 to 1 with increments of 0.001. 
Create a list of all Ridge Regression R^2 scores for training and testing data. 


````python
# Write your code below and press Shift+Enter to execute 

# iterate over a range of alpha values for Ridge regularization
Alpha = np.linspace(0.001, 1, 1000)

Rsqu_test = []
Rsqu_train = []

pbar = tqdm(Alpha, ncols=100)

for alpha in pbar:
    # instantiate a Ridge Regression model withh current alpha value
    RidgeModel = Ridge(alpha=alpha)

    # fit the model on the polynomial-transformed training data
    RidgeModel.fit(X_train_pr, y_train)

    # compute the R^2 scores for both the training and test sets
    test_score, train_score = RidgeModel.score(X_test_pr, y_test), RidgeModel.score(X_train_pr, y_train)

    # update the progress bar with the current scores
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    # store the R^2 score
    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
````

Plot the R^2 values for training and testing sets with respect to the value of alpha


````python
# Write your code below and press Shift+Enter to execute 
plt.figure(figsize=(10, 6))  
plt.plot(Alpha, Rsqu_test, label='validation data')
plt.plot(Alpha, Rsqu_train, 'r', label='training Data')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.ylim(0.3, 0.8)
plt.legend();
````
